# FUNCIONES DE AGRUPACIÓN

```python
import pandas as pd
df_books = pd.read_csv('https://github.com/cesarmayta/datasets/raw/refs/heads/main/libros.csv')
df_books.head(2)
```

# GROUP BY

```python
df_books.groupby('Author').sum().reset_index()
```

```python
df_books.groupby('Author')[['User Rating','Reviews','Price']].mean().reset_index()
```

## AGREGACIONES

```python
cols = ['User Rating','Reviews','Price']
```

```python
df_books.groupby('Author')[cols].agg(['min','max'])
```

```python
df_books.groupby('Author')[cols].agg({'Reviews':['min','max'],'User Rating':'sum','Price':'mean'}).reset_index()
```

```python
df_books.groupby('Author')[cols].agg(rating_maximo=('User Rating','max'),precio_promedio=('Price','mean')).reset_index()
```

# AGRUPACIÓN POR 2 COLUMNAS

```python
df_books.groupby(['Year','Genre'])[cols].count().reset_index()
```
